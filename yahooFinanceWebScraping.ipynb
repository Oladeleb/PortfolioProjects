```python
import requests
from bs4 import BeautifulSoup
```

```python
url= 'https://finance.yahoo.com/quote/AAPL?p=AAPL&.tsrc=fin-srch'
```

```python
response = requests.get(url)
print(response)
```

```python
soup =BeautifulSoup(response.text, 'lxml').head()
print(soup)
```

```python
classtag = soup.find_all('div', {'D(ib) Mend(20px)'}).head()
print(classtag)
```

```python
price = soup.find_all ('div', {'D(ib) Mend(20px)'})[0].find('fin-streamer').text
print(price)
```

```python
# Call a function that retrieve  AAPL price from Yahoo finance
def priceTracker():
    url= 'https://finance.yahoo.com/quote/AAPL?p=AAPL&.tsrc=fin-srch'
    response = requests.get(url)
    soup =BeautifulSoup(response.text, 'lxml')
    price = soup.find_all ('div', {'D(ib) Mend(20px)'})[0].find('fin-streamer').text
    return price
   
while True:
    print ('Current Price of Apple:'+ priceTracker())
```
